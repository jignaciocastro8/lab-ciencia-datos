### 쯏a, pero esto tiene aplicaciones m치s all치 de guardar informaci칩n en casilleros?

Para el caso de im치genes podemos encontrar m칰ltiples aplicaciones con la manipulaci칩n de los tensores y operando matem치ticamente con ellos. Algunas de las aplicaciones m치s conocidas (y que aplicaremos) son las siguientes:

- **Obtener el negativo de una imagen**: Consiste en obtener el color complementario de una imagen, para esto debemos restar la imagen con la m치xima intensidad que podemos encontrar en una imagen (o sea 255).
    
    
- **Pasar a escala de grises una imagen**: Los valores RGB se convierten a escala de grises mediante la f칩rmula NTSC: 

    $$ imagen\_gris = 0.299 * Rojo + 0.587 * Verde + 0.114 * Azul $$
    
    Esta f칩rmula representa la percepci칩n relativa de la persona promedio del brillo de la luz roja, verde y azul.
    
    
- **Mejora de contraste**: Son m칰ltiples las t칠cnicas que nos permiten mejorar el contraste de una imagen, pero, una t칠cnica simple para modificar los contrastes consta en obtener un factor de correcci칩n llamado F en base al contraste deseado (C). Luego, es aplicado en la diferencia entre la imagen y 128. De esta forma obtenemos R, que es la imagen con la mejora de contraste deseada.

    $$ F=259*(C+255)/(255*(259-C)) $$
    $$ R=F*(img-128)+128 $$
  
  
- **Convoluci칩n**: Consiste en recorrer una imagen por cada uno de sus canales utilizando una matriz que lleva por nombre Kernel. El kernel, examinar치 los conjuntos de pixeles que recorre, aplicando una multiplicaci칩n de los valores circundantes ,y sumando todos los valores generados de este producto para generar un nuevo pixel en el tensor de salida.

- La idea es que la imagen contenida en la clase sea inmutable, por ende, todos los metodos que modifiquen la imagen contenida en el objeto deberan retornar un nuevo objeto de la clase `Imagen` que contenga la imagen modificada.

````python
# Numpy maneja las exception de dimensionalidad de los arreglos.

class Imagen:
    """Clase contenedora de im치genes. Contiene como atributo
    un arreglo np.array de 3 dimensiones."""
    
    def __init__(self, img):
        if isinstance(img, np.ndarray):
            if img.ndim != 3:
                raise ("El argumento debe ser un arreglo de numpy de solo 3 dimensiones")
            if img.shape[-1] != 3:
                raise (
                    "El argumento debe ser un arreglo de numpy de solo 3 dimensiones "
                    "tal que la 칰ltima dimensi칩n solo tiene 3 canales"
                )
            # Cast to int.
            self.data = img.astype(int)
        else:
            raise TypeError(
                "Debes entregar un arreglo de numpy como argumento del constructor de "
                "Imagen"
            )

    def show(self):
        """Muestra la im치gen contenida en la propiedad imagen usando plotly.
        Lo mismo que la funci칩n mostrar_imagen.
        """
        plt.imshow(self.data)
        plt.show()
        x, y, z = self.data.shape
        print(f'Tama침o en cada dimensi칩n: ({x}, {y}, {z})')
        pass

    def info(self):
        """ Imprime las caracter칤sticas de la imagen cargada: Alto y ancho.
        """
        print(f'Alto (# filas): {self.data.shape[0]} \nAncho (# columnas): {self.data.shape[1]}')
        pass

    def __add__(self, other):
        """Redefine la operaci칩n + entre imagen y escalar (imagen + other).

        # Idea, usar indexado condicial (similar a los filtros de pandas).
        # Sumar y luego que en cada pixel mayor a 255 sea asignado el m치ximo.
        # Ver los tests para mas informaci칩n.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        # int or float.
        if isinstance(other, float) or isinstance(other, int) or isinstance(other, np.ndarray):
            result = (self.data + other).astype(int)
            # Saturar.
            result[result > 255] = 255
            result[result < 0] = 0  
            return Imagen(result)
        # numpy array.
        if isinstance(other, Image):
            result = self.data + other.data
            result[result > 255] = 255
            result[result < 0] = 0  
            return Imagen(result)
        else:
            raise TypeError("Debe entregar un arreglo numpy, float, int or Imagen.")

    def __radd__(self, other):
        """Operaci칩n conmutativa de __add__. (other + imagen).

        Hint: deber칤a llamar a __add__...

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        return self + other


    def __sub__(self, other):
        """Redefine la operaci칩n + entre imagen y escalar. (image - other)

        # Idea, usar indexado condicial (similar a los filtros de pandas).
        # Restar y luego que en cada pixel mayor a 255 sea asignado el m치ximo.
        # Ver los tests para mas informaci칩n.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        # int or float.
        if isinstance(other, float) or isinstance(other, int) or isinstance(other, np.ndarray):
            result = (self.data - other).astype(int)
            # Saturar.
            result[result < 0] = 0
            result[result > 255] = 255  
            return Imagen(result)
        # numpy array.
        if isinstance(other, Imagen):
            result = self.data - other.data
            result[result < 0] = 0
            result[result > 255] = 255  
            return Imagen(result)
        else:
            raise TypeError("Debe entregar un arreglo numpy, float or int.")

    def __rsub__(self, other):
        """Operaci칩n conmutativa de __sub__. (other - Image)

        Transform other to an Image object.

        Hint: deber칤a llamar a __sub__...

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        # int or float.
        if isinstance(other, float) or isinstance(other, int) or isinstance(other, np.ndarray):
            # create image.
            other_img = (other * np.ones(self.data.shape)).astype(int)
            return Imagen(other_img) - self
        # numpy array.
        if isinstance(other, np.ndarray):
            return Imagen(other.astype(int)) - self
        else:
            raise TypeError("Debe entregar un arreglo numpy, float or int.")


    def __mul__(self, other):
        """Redefine la operaci칩n + entre imagen y escalar. (image * other)

        # Idea, usar indexado condicial (similar a los filtros de pandas).
        # Sumar y luego que en cada pixel mayor a 255 sea asignado el m치ximo y 
        # cada valor inferior a 0 debe ser 0.
        # Ver los tests para mas informaci칩n.

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        # int or float.
        if isinstance(other, float) or isinstance(other, int) or isinstance(other, np.ndarray):
            result = (self.data * other).astype(int)
            # Saturar.
            result[result < 0] = 0
            result[result > 255] = 255
            return Imagen(result)
        # numpy array.
        if isinstance(other, Image):
            result = self.data * other.data
            result[result < 0] = 0
            result[result > 255] = 255 
            return Imagen(result)
        else:
            raise TypeError("Debe entregar un arreglo numpy, float or int.")

    def __rmul__(self, other):
        """Operaci칩n conmutativa de __mul__. (other * image)

        Hint: deber칤a llamar a __mul__...

        Parameters
        ----------
        other : Union[int, np.ndarray]
            Escalar o arreglo que ser치 sumado a cada pixel de la imagen
        """
        return self * other
    
````

````python
class ProcImage():
    @staticmethod
    def negative(img_in):
        """Makes img_in negative.

        Args:
            img_in (Imagen): Image object.
        """
        return 255 - img_in
    
    @staticmethod
    def img2gray(img_in):
        """Turns img_in into gray scale image.

        Args:
            img_in (Imagen): Imagen object.
        """
        arr = 0.299 * img_in.data[:,:,0] + 0.587 * img_in.data[:,:,1] + 0.114 * img_in.data[:,:,2]
        return Imagen(np.stack([arr, arr, arr], axis=2))
    
    @staticmethod
    def imgs2gray(imgs_in):
        """Array version of img2gray().

        Args:
            imgs_in (Array): Array of Imagen objects.
        """
        return [ProcImage.img2gray(img) for img in imgs_in]
  
    @staticmethod
    def img2RGB(img_in, color):
        """Returns the channel of color "color".

        Args:
            img_in (Imagen): Imagen object.
            color (str): Channel. Must be 'r', 'g' or 'b'.
        
        Returns:
            (): 
        """
        color_idx = {'r' : 0, 'g' : 1, 'b' : 2}
        out = np.zeros(img_in.data.shape)
        out[:,:,color_idx[color]] = img_in.data[:, :, color_idx[color]]
        return Imagen(out)

    @staticmethod
    def img2contrast(img_in, C):
        """Apply a level of constrast "C" to "img_in".

        Args:
            img_in (Imagen): Imagen object.
            C (float): Level of contrast.
        """
        F = 259 * (C + 295) / 255 * (259 - C)
        return F * (img_in - 128) + 128
    
    @staticmethod
    def conv_channel(img_in, kernel):
        """Apply the convolution process on img_in.

        Args:
            img_in (Imagen): Imagen object.
            kernel (2D array): Array to convolve with.

        Returns:
            [Imagen]: Convolved image.
        """
        img = img_in.data
        img_out = []
        for i in range(img.shape[-1]):
            img_channel = convolve2d(img[:, :, i], 
                                     kernel, 
                                     mode="same", 
                                     boundary="symm")
            img_out.append(img_channel)
        return Imagen(np.stack(img_out, axis=2).astype(int))
````
